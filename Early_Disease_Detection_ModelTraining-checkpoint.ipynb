```python
def bmi(weight, height):
    height = height / 100
    ans = weight / (height * height)
    return ans
```

```python
def isRiskBMI(bmi):
    if bmi < 18.5:
        return "Under Weight"
    elif bmi >= 18.5 and bmi <= 24.9:
        return "Normal"
    elif bmi > 24.9 and bmi <= 29.9:
        return "Overweight"
    else:
        return "Obesity"
```

```python
def sleep_risk(sleep_hour):
    isRisky = 1 if sleep_hour < 6 else 0
    return isRisky
```

```python
def metabolic_risk(activity, screen_time):
    sedentary_risk = 1 if (activity < 4 and screen_time > 7) else 0
    return sedentary_risk
```

```python
def dehyrdation_risk(water_glass, hydration):
    water_risk = 1 if (water_glass < 5 or hydration < 4) else 0
    return water_risk
```

```python
def vitaminD_risk(sun_exposer):
    sun_risk = 1 if sun_exposer < 3 else 0
    return sun_risk
```

```python
df["sleep_risk"] = (df["sleep_hour"] < 6).astype(int)
df["sedentary_risk"] = ((df["activity"] < 4) & (df["screen_time"] > 7)).astype(int)
df["dehydration_risk"] = ((df["water_glass"] < 5) | (df["hydration"] < 4)).astype(int)
df["sun_risk"] = (df["sun_exposer"] < 3).astype(int)
```

```python
df["metabolic_risk_score"] = df["bmi"] + df["stress"] + df["fatigue"] - df["activity"]
df["lifestyle_risk_score"] = df["screen_time"] + df["stress"] - df["sleep_hour"]
```

```python
df["overall_risk_score"] = (
    df["metabolic_risk_score"] +
    df["lifestyle_risk_score"] +
    df["sleep_risk"] +
    df["sedentary_risk"] +
    df["dehydration_risk"] +
    df["sun_risk"]
)
```

```python
plt.figure()
plt.plot(df.index, df["overall_risk_score"])
plt.xlabel("Individual Index")
plt.ylabel("Overall Risk Score")
plt.title("Silent Disease Risk Score Progression")
plt.show()
```

```python
def predict_disease_risk(model_path, input_features):
    """
    model_path: path to .pkl model
    input_features: list of feature values in order
    """
    with open(model_path, "rb") as f:
        model = pickle.load(f)

    input_array = np.array(input_features).reshape(1, -1)
    # Predicting probability
    if hasattr(model, "predict_proba"):
        risk = model.predict_proba(input_array)[0][1] * 100
    else:
        risk = model.predict(input_array)[0] * 100

    return round(risk, 2)
```
